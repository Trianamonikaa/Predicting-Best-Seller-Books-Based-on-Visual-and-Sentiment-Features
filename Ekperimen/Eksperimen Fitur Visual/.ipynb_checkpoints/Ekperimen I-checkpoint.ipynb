```python
from sklearn.svm import SVC
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import train_test_split, KFold
from sklearn.cross_validation import cross_val_score,cross_val_predict
from imutils import paths
import numpy as np
import imutils
import os
import pandas as pd
import cv2
from sklearn.metrics import confusion_matrix,accuracy_score 
from sklearn.metrics import classification_report
```

```python
def image_to_feature_vector(image, size=(50, 50)):
    return cv2.resize(image, size).flatten()
RGB = []
labels = []
images = []
imagePaths = list(paths.list_images("data_gambar"))
for (i, imagePath) in enumerate(imagePaths):
    image = cv2.imread(imagePath)
    images.append(image)
    label = imagePath.split(os.path.sep)[-1].split(".")[0]
    rgb = image_to_feature_vector(image)
    labels.append(label)
    RGB.append(rgb)  

RGBfeatures = np.array(RGB)
labels = np.array(labels)

print("[INFO] rgb matrix: {:.2f}MB".format(RGBfeatures.nbytes / (1024 * 1000.0)))
```

```python
mattrix_image = pd.DataFrame(RGBfeatures,labels)
mattrix_image
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits = 10)
for train, test in kf.split(RGBfeatures):
    X = RGBfeatures 
    y =  labels
    print (train)
    print (test)
    X_Train = X[train]
    X_test = X[test]
    y_train = y[train]
    y_test = y[test]
    print( X_Train)
    print(X_test)
    print(y_train)
    print(y_test)
    X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]
    print("Size of X training set: {} size of X test set: {} size of Y training set: {} size of Y test set: {} \n".format(X_train.shape, X_test.shape, y_train.shape, y_test.shape))
```

```python
svm_rgb_model = SVC()
model_RGB = svm_rgb_model.fit(RGBfeatures,labels)
```

### Cross Validation 

```python
print(cross_val_score(svm_rgb_model , RGBfeatures, labels, scoring='accuracy', cv = 10))
accuracy_cv10 = cross_val_score(model_RGB , RGBfeatures, labels, scoring='accuracy', cv = 10).mean()
print("akurasi RGB feature untuk K10", accuracy_cv10)
```
