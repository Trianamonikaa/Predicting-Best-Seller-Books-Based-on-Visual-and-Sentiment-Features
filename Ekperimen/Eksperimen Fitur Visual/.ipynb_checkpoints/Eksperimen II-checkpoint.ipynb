```python
from sklearn.svm import SVC
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import train_test_split, KFold
from sklearn.cross_validation import cross_val_score,cross_val_predict
from imutils import paths
import numpy as np
import imutils
import os
import pandas as pd
import cv2
from sklearn.metrics import confusion_matrix,accuracy_score 
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
def image_to_hsv_feature_vector(image, size=(50, 50)):
    rimage = cv2.resize(image, size)
    return cv2.cvtColor(rimage, cv2.COLOR_BGR2HSV).flatten()

HSV = []
labels = []

imagePaths = list(paths.list_images("data_gambar"))
for (i, imagePath) in enumerate(imagePaths):
    image = cv2.imread(imagePath)
    label = imagePath.split(os.path.sep)[-1].split(".")[0]
    hsv = image_to_hsv_feature_vector(image)
    HSV.append(hsv)
    labels.append(label)
    

HSVfeatures = np.array(HSV)
labels = np.array(labels)
print("[INFO] HSV features matrix:{:.2f}MB".format(HSVfeatures.nbytes / (1024 * 1000.0)))
```

```python
mattrix_image = pd.DataFrame(HSVfeatures ,labels)
mattrix_image
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10)
for train, test in kf.split(HSVfeatures):
    X = HSVfeatures 
    y =  labels
    print (train)
    print (test)
    X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]
    print("Size of X training set: {} size of X test set: {}\n".format(X_train.shape, X_test.shape))
#     print("Size of Y training set: {} size of Y test set: {}\n".format(y_train1.shape, y_test1.shape))
```

```python
svm_hsv_model = SVC()
HSV_model = svm_hsv_model.fit(HSVfeatures, labels)
```

### Cross Validation 

```python
print(cross_val_score(svm_hsv_model, HSVfeatures, labels, scoring='accuracy', cv = 10))
accuracy_cv10 = cross_val_score(svm_hsv_model,HSVfeatures, labels, scoring='accuracy', cv = 10).mean()
print("akurasi HSV feature untuk K10", accuracy_cv10)
```
