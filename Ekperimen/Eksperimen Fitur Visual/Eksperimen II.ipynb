```python
from sklearn.svm import SVC
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import train_test_split, KFold
from sklearn.cross_validation import cross_val_score,cross_val_predict
from imutils import paths
import numpy as np
import imutils
import os
import pandas as pd
import cv2
from sklearn.metrics import confusion_matrix,accuracy_score 
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
def image_to_hsv_feature_vector(image, size=(50, 50)):
    rimage = cv2.resize(image, size)
    return cv2.cvtColor(rimage, cv2.COLOR_BGR2HSV).flatten()

HSV = []
labels = []

imagePaths = list(paths.list_images("data_gambar"))
for (i, imagePath) in enumerate(imagePaths):
    image = cv2.imread(imagePath)
    label = imagePath.split(os.path.sep)[-1].split(".")[0]
    hsv = image_to_hsv_feature_vector(image)
    HSV.append(hsv)
    labels.append(label)
    

HSVfeatures = np.array(HSV)
labels = np.array(labels)
print("[INFO] HSV features matrix:{:.2f}MB".format(HSVfeatures.nbytes / (1024 * 1000.0)))
```

```python
category_encoder = LabelEncoder()
y = category_encoder.fit_transform(labels)
```

```python
mattrix_image = pd.DataFrame(HSVfeatures ,labels)
mattrix_image
```

```python
C = 1.0  # SVM regularization parameter
svm_hsv_model = SVC(probability=True, kernel='linear', C=C)
# model_RGB = svm_rgb_model.fit(RGBfeatures,labels)
```

### Evaluasi fitur visual 

```python
p = cross_val_score(svm_hsv_model,HSVfeatures, y, cv=10, scoring='precision')
print("precision:", sum(p)/len(p))
r = cross_val_score(svm_hsv_model,HSVfeatures, y, cv=10, scoring='recall')
print("recall:", sum(r)/len(r))
f = cross_val_score(svm_hsv_model,HSVfeatures, y, cv=10, scoring='f1')
print("f1:", sum(f)/len(f))
```
