```python
from sklearn.svm import SVC
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import train_test_split, KFold
from sklearn.cross_validation import cross_val_score,cross_val_predict
from imutils import paths
import numpy as np
import imutils
import os
import pandas as pd
import cv2
from sklearn.metrics import confusion_matrix,accuracy_score 
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
def image_to_feature_vector(image, size=(50, 50)):
    return cv2.resize(image, size).flatten()
RGB = []
labels = []
images = []
imagePaths = list(paths.list_images("data_gambar"))
for (i, imagePath) in enumerate(imagePaths):
    image = cv2.imread(imagePath)
    images.append(image)
    label = imagePath.split(os.path.sep)[-1].split(".")[0]
    rgb = image_to_feature_vector(image)
    labels.append(label)
    RGB.append(rgb)  

RGBfeatures = np.array(RGB)
labels = np.array(labels)

print("[INFO] rgb matrix: {:.2f}MB".format(RGBfeatures.nbytes / (1024 * 1000.0)))
```

```python
category_encoder = LabelEncoder()
y = category_encoder.fit_transform(labels)
```

```python
mattrix_image = pd.DataFrame(RGBfeatures,labels)
mattrix_image
```

```python
C = 1.0  # SVM regularization parameter
svm_rgb_model = SVC(probability=True, kernel='linear', C=C)
# model_RGB = svm_rgb_model.fit(RGBfeatures,labels)
```

### Evaluasi fitur visual

```python
p = cross_val_score(svm_rgb_model,RGBfeatures, y, cv=10, scoring='precision')
print("precision:", sum(p)/len(p))
r = cross_val_score(svm_rgb_model,RGBfeatures, y, cv=10, scoring='recall')
print("recall:", sum(r)/len(r))
f = cross_val_score(svm_rgb_model,RGBfeatures, y, cv=10, scoring='f1')
print("f1:", sum(f)/len(f))
```
